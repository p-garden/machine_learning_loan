## Neural Network를 활용한 회귀 문제 예시

## 1. 데이터셋 만들기

```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt

X_train = np.linspace(-10, 10, 1000)
y_train = np.sin(X_train) + np.random.normal(0, 0.2, size=X_train.shape)

X_test =  np.linspace(-10, 10, 500)
y_test =  np.sin(X_test) + np.random.normal(0, 0.2, size=X_test.shape)
```

```python
plt.scatter(X_train, y_train)
plt.show()
```

## 2. 선형회귀

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X_train.reshape(-1,1), y_train)
y_pred = slr.predict(X_test.reshape(-1,1))
plt.plot(X_test,y_pred,color='red')
plt.scatter(X_test,y_test)
```

## 3.MLP Regressor

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
regr = MLPRegressor(random_state=1, max_iter=500,hidden_layer_sizes=(62,62),activation='relu').fit(X_train.reshape(-1,1), y_train)
# 히든 층의 노드 (62)로 지정
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, regr.predict(X_test.reshape(-1,1)),color='red')
plt.show()
```
