## 2. 데이터프레임 생성

```python
# 라이브러리를 임포트합니다.
import pandas as pd

# 데이터프레임을 만듭니다.
dataframe = pd.DataFrame()

# 열을 추가합니다.
dataframe['Name'] = ['Jacky Jackson', 'Steven Stevenson']
dataframe['Age'] = [38, 25]
dataframe['Driver'] = [1, 0]

# 데이터프레임을 확인합니다.
dataframe
```

```python
# 행을 만듭니다.
new_person = pd.Series(['Molly Mooney', 40, 1], index=['Name','Age','Driver'])

# 행을 추가합니다.
dataframe.append(new_person, ignore_index=True)
```

```python
# 행을 만듭니다.
new_person = pd.Series(['Molly Mooney', 40, 1], index=['Name','Age','Driver'])

# 행을 추가합니다.
pd.concat([dataframe,new_person.to_frame().T],ignore_index=True)
#concat는 데이터프레임끼리 이어붙이기만 가능 -> 새 series를 데이터프레임으로 변환 후 이어 붙이기
```

```python
# 행 바로 추가
dataframe.loc[len(dataframe)]=["aa",50, 0]
dataframe
```

```python
# 라이브러리를 임포트합니다.
import pandas as pd


# 데이터프레임으로 데이터를 적재합니다.
dataframe = pd.read_csv('data/titanic.csv')

# 첫 번째 행을 선택합니다.
dataframe.iloc[0]
```

### 조건에 따라 행 선택

```python
# 통곗값을 계산합니다.
print('최댓값:', dataframe['Age'].max())
print('최솟값:', dataframe['Age'].min())
print('평균:', dataframe['Age'].mean())
print('합:', dataframe['Age'].sum())
print('카운트:', dataframe['Age'].count())
```

### Nan 값 처리

```python
import numpy as np
#random한 3행 4열의 DataFrame 생성
data_frame = pd.DataFrame(np.random.randn(3, 4))
data_frame
```

```python
#data_frame의 모든 원소에 log 값을 취함
new_data_frame = np.log(data_frame)
#log씌워서 음수값 Nan만들기
```

```python
new_data_frame
```

```python
# nan 값이 있는 행 제거
new_data_frame.dropna()
# Nan이 하나라도 있는 행은 제거
```

```python
# 0 열에 nan값이 있는 행 제거
new_data_frame.dropna(subset=[0])
```

```python
# 1행에 nan 값이 있는 열 제거
new_data_frame.dropna(axis=1,subset=[1])
```

```python
#모든 NaN을 0으로 치환한다
new_data_frame.fillna(0)
```

```python
#NaN을 같은 열의 바로 위의 행 값으로 대체
new_data_frame.fillna(method='ffill')
```

```python
#NaN을 같은 열의 바로 아래 행 값으로 대체
new_data_frame.fillna(method='bfill')
```
